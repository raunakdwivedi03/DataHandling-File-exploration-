```python
class MyException(Exception):
  def __init__(self,message):
    print(message)

class Bank:

    def __init__(self, balance):
        self.balance = balance

    def withdraw(self, amount):

        if amount < 0:
            raise MyException('invalid amount, cannot be negative')

        if amount > self.balance:
            raise MyException('insufficient balance')

        self.balance = self.balance - amount


obj = Bank(10000)

try:
    obj.withdraw(599)
    print(obj.balance)

except MyException as e:
    print(e)

else:
    print(obj.balance)
```

```python
class SecurityError(Exception):

    def __init__(self, message):
        print(message)

    def logout(self):
        print('logout')


class Google:

    def __init__(self, name, email, password, device):
        self.name = name
        self.email = email
        self.password = password
        self.device = device

    def login(self, email, password, device):
        if device != self.device:
            raise SecurityError('bhai tere to lag gye')

        if email == self.email and password == self.password:
            print('welcome')
        else:
            print('login error')


obj = Google('raunak', 'raunak@gmail.com', '1234', 'andriod')

try:
    obj.login('raunak@gmail.com', '1234', 'andriod')

except SecurityError as e:
    e.logout()

else:
    print(obj.name)

finally:
    print('database connection closed')
```
