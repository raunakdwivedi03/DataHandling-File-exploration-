#Decorators


A decorator in python is a funtion that recieve another function as input and adds some functionality(decoration) to and it and return it.
This can happens only because phython function are 1st clss citizen.


There are 2 types decorators avialable
in python

-->Build in decorators like @staticmethod,@classmethod , @abstractmethod , and @property etc.


-->user defined decorators that we programers can create according to our needs.


```python
#python are 1st class function
def modify(func,num):
  return func(num)

def square(num):
  return num**2

print(modify(square,5))
```

```python
##simple example
def  my_decorator(func):
  def wrapper():
    print('before')
    func()
    print('after')
  return wrapper


def hello():
  print('hello')

def display():
  print('hello raunak')

a=my_decorator(hello)
a()

b=my_decorator(display)
b()
```
