using context manager(with)


```python
## with
with open('/content/sample1.txt','w') as f:
  f.write('salmon bhai ')

  ## this is the with method in which the file will close automatically when work is done
  #there is no need to write f.close( ) or any thing
```

```python
## try f.read() now
with open('/content/sample.txt','r') as f:
  print(f.readline())

```

```python
## moving with in a file -> 10 char then 10 char
with open('/content/sample.txt','r') as f:
  print(f.read(10))
  print(f.read(10))
```

```python
## benefit ? to load a big file in memory
big_l=['hello world' for i in range(1000)]

with open('big.txt','w') as f:
  f.writelines(big_l)
```

```python
## loading file chhunck by chuncks
with open('big.txt','r') as f:
  chunk_size=100

  while len(f.read(chunk_size)) > 0:
    print(f.read(chunk_size),end='***')
    f.read(chunk_size)


## this is the best way to read and process the big files
```

```python
## seek and tell functions
with open('sample.txt','r') as f:
  print(f.read(10))
  print(f.tell()) ## f. tell is ,mechanism to tell the 10 char and moving toward 11th char

  f.seek(4) ## seek vaps starting point pe le aata hai mtlb khi bhii ja skte hoo
  print(f.read(10))
  f.tell()

```

```python
##seek during write
with open('sample.txt','w') as f:
  f.write('hello')
  f.seek(0)
  f.write('xaal')
```

**Serialization & Deserialisation**



*   Serialization-> process of converting python data type to JSON format

*  eserialization-> process of converting JSON
to python data type



JSON --> JSON ka full form hai JavaScript Object Notation.

JSON kya hota hai? (Hinglish)

JSON ek lightweight data format hai jo data store aur data transfer karne ke liye use hota hai.
Mostly server aur client ke beech data bhejne ke liye use hota hai.

```python
#JSON key–value pair mein hota hai, bilkul Python dictionary jaisa
{
  "name": "Raunak",
  "age": 22,
  "is_student": True
}
```

```python
##serialization
#list

import json
l=[1,2,3,4,5,]

with open('demo.json','w') as f:
  json.dump(l,f)
```

```python
#dictonary
d={
  "name": "Raunak",
  "age": 22,
  "is_student": True
}

with open('demo.json','w') as f:
  json.dump(d,f,indent=4)
```

```python
##deserialization
import json

with open('demo.json','r') as f:
  d=json.load(f)
  print(d)
  print(type(d))
```

```python
#serialise and deserialise tuple
import json

t=(1,2,3,4,5,6)

with open('demo.json','w') as f:
  json.dump(t,f)
```

```python
#serialize and deserialize a nested dict
d={
    'student':'raunak',
    'marks':[23,4,2,3,2,23]
}

with open('demo1.json','w') as f:
  json.dump(d,f)
```

Serializing and Deserializing custom objects


```python
class Person:

  def __init__(self,fname,lname,age,gender):
    self.fname=fname
    self.lname=lname
    self.age=age
    self.gender=gender

## mujhe es format me chahie
# raunak -> dwiveddi -> 22-> male
```

```python
person=Person('raunak','dwivedi',22,'male')
```

```python
# as a string
import json

def show_obj(person):
  if isinstance(person,Person):
    return "{} {} age -> {} gender -> {}".format(person.fname,person.lname,person.age,person.gender)

with open('demo2.json','w') as f:
  json.dump(person,f,default=show_obj)
```

Pickling -> Pickling Python ka ek process hai jisme Python object ko file mein save kar dete hain binary format mein.
Is process ko bolte hain Serialization.

Matlab:

Object → File (save karna) = Pickling

File → Object (wapis lana) = Unpickling

```python
class Person:

  def __init__(self,name,age):
    self.name=name
    self.age=age

  def display_info(self):
    print('hi my name is ',self.name,'and i am', self.age,'year old')
```

```python
p=Person('raunak ' , 22)
```

```python
##pickling
#pickle dump
import pickle

with open('person.pkl','wb') as f:
  pickle.dump(p,f)
```

```python
##pickle load
import pickle
with open('person.pkl','rb') as f:
  p=pickle.load(f)

p.display_info()
```

Pickle Vs Json

--pickle lets the user to store data in binary format.Json lets the user store the data in a human readable text format.
